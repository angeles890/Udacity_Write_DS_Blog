### Data Review Notes & To Do
</br>
<ul>
    <li>1. Date data stored as string, must convert to DateTime to support time series analysis</li>
    <li>2. Some tables have columns were all values are missing (ie 'jurisdiction_names' in Listings), remove such columns</li>
    <li>3. Price data is stored as a string, need to convert to float/decimal</li>
    <li>4. Add EOMonth type column to allow aggregation by month</li>
</ul>

```python
# 1. Convert date from string to date

# boston Data
df_boston_calendar['date'] = pd.to_datetime(df_boston_calendar['date'])
df_boston_reviews['date'] = pd.to_datetime(df_boston_reviews['date'])
df_boston_listings['last_scraped'] = pd.to_datetime(df_boston_listings['last_scraped'])

```

```python
# 2. remove columns that are missing all values
listings_boston_na = set(df_boston_listings.loc[:,df_boston_listings.isna().sum()/df_boston_listings.shape[0]==1.0].columns)
df_boston_listings.dropna(axis=1,how='all',inplace = True)

assert len(listings_boston_na.intersection(set(df_boston_listings.columns))) == 0
```

```python
# 3. Convert string currency to float/decimal data
df_boston_calendar['price'] = df_boston_calendar['price'].fillna('$0.00')
df_boston_listings['price'] = df_boston_listings['price'].fillna('$0.00')
df_boston_listings['weekly_price'] = df_boston_listings['weekly_price'].fillna('$0.00')
df_boston_listings['monthly_price'] = df_boston_listings['monthly_price'].fillna('$0.00')

def convert_str_currency(val):
    '''
        Input: Currency in string format, such as "$65.50"
        Output: Float value of inputted number 65.50 (Float)
    '''
    
    value = sub(r'[^\d.]','',val)
    return value

df_boston_calendar['price'] = df_boston_calendar['price'].apply(convert_str_currency).astype(float)
df_boston_listings['price'] =  df_boston_listings['price'].apply(convert_str_currency).astype(float)
df_boston_listings['weekly_price'] = df_boston_listings['weekly_price'].apply(convert_str_currency).astype(float)
df_boston_listings['monthly_price'] = df_boston_listings['monthly_price'].apply(convert_str_currency).astype(float)


```

```python
# 4. Create EOMonth Column to support data analysis across type
def get_eomonth(val):
    eomonth = val + MonthEnd(0)
    return eomonth
df_boston_calendar['eomonth'] = df_boston_calendar['date'].apply(get_eomonth)
df_boston_reviews['eomonth'] =  df_boston_reviews['date'].apply(get_eomonth)
df_boston_listings['eomonth'] = df_boston_listings['last_scraped'].apply(get_eomonth)
```

### Question 1
> How do seasonality impact pricing and inventory within Boston?

```python
import matplotlib.dates as mdates
# create plot object
plt.figure(figsize = [15,15])

# create custom ticks and labels 
# cretae a list set to the unique months in our dataset
x_ticks = range(0,df_boston_calendar['eomonth'].nunique())
# split the date and take only first part in order remove UTC time data from date
x_labels = [str(m).split('T')[0] for m in np.sort(df_boston_calendar['eomonth'].unique())]

# Plot bar chart with average listing price each month
#plt.subplot(1,2,1)
# filter for only items listed as availiable
df_boston_calendar.query('available == "t"').groupby('eomonth')['price'].mean().plot(kind='bar',x='eomonth',y='listing_id');
plt.title('Median Listing Price by Month',fontsize=24)
plt.ylabel('Median Listing Price ($)',fontsize=18)
plt.yticks(fontsize=16)
plt.xlabel('Month',fontsize=18);
plt.xticks(x_ticks,x_labels, fontsize=16)
# ensures all the labels are in the resulting png file
plt.tight_layout()
# save figure as png
plt.savefig('median_listing_price_by_month.png')



```

From the above chart showing median price per month, it's clear that seasonality plays a large impact on pricing. The cheapest months also coincide with the coldes months in Boston, which makes sense as there would be less demand from outsiders. 

<strong>Inventory</strong>
A simple way to asses inventory is to count the number of listings listed as availiable each month, however, we need to account for the variation in the number of days in different months. For example, February has less days than July and as a result should have more listings availiable. To control for this, we will analyze the average number of listings availiable per day in each month by getting the total count of availiable listings per month and dividing it by the number of days in that month.

```python
# get number of unique days in each month in our dataset where there is at least 1 listing availiable
df_unique_days_per_month = df_boston_calendar.query('available == "t"').groupby('eomonth')['date'].nunique().to_frame().reset_index()
# rename column
df_unique_days_per_month.rename(columns={'date':'days_cnt'},inplace=True)
# for each month, get count of avaliable listings
df_listings_per_month = df_boston_calendar.query('available == "t"').groupby('eomonth')['listing_id'].count().to_frame().reset_index()
# rename column
df_listings_per_month.rename(columns={'listing_id':'listing_cnt'},inplace=True)

df_listings_per_month = df_listings_per_month.merge(df_unique_days_per_month,how='inner',left_on='eomonth',right_on='eomonth')
df_listings_per_month['avg_per_day'] = df_listings_per_month['listing_cnt']/df_listings_per_month['days_cnt']
# create plot object
#plt.figure(figsize = [20,20])

# plot bar chart with count of listings each month
#plt.subplot(1,2,2)
# filter for only items listed as availiable
#df_boston_calendar.query('available == "t"').groupby('eomonth')['listing_id'].nunique().plot(kind='bar',x='eomonth',y='listing_id');
df_listings_per_month.plot(kind='bar',x='eomonth',y='avg_per_day',figsize=(15,15));
# set title
plt.title('Average Daily Listings by Month',fontsize=24);
# set y label and ticks
plt.ylabel('Average Daily Listing Count',fontsize=18);
plt.yticks(fontsize=16);
# set x label and ticks
plt.xlabel('Month',fontsize=18);
plt.xticks(x_ticks, x_labels,fontsize=16);
# set layout as tight to ensure labels and title show up in png
plt.tight_layout()
# save as png
plt.savefig('cnt_unqiue_listings_by_month.png');
```

In the above chart, the greatest level of inventory exists between November and February, which overlaps with the period of lowest prices and period of coldest weather. Again this makes sense as less visitors to Boston will result in a higher number of units not being booked and thus being available.

<strong>Modeling Price</strong>
<br>
Using data from the listing data frame, and insights gleaned from the previous visiuals a simple linear model will be developed to predict listing price.

```python
# init new list to hold new column names
new_columns = []
# encode categorical variables
df_neighbourhood_boston = pd.get_dummies(df_boston_listings.neighbourhood_cleansed,drop_first = True,prefix_sep='neighborhood_')
# for each column replace ' ' with '_'
for c in list(df_neighbourhood_boston.columns):
    col = c.replace(' ','_')
    # append 'neighborhood_' to each column name
    col = 'neighborhood_'+col
    # add to new columns list
    new_columns.append(col)

# set new column names
df_neighbourhood_boston.columns = new_columns

new_columns = []
# encode categorical variables
df_room_type_boston = pd.get_dummies(df_boston_listings.room_type, drop_first=True,prefix_sep='room_type_')
# for each column replace ' ' with '_'
for c in list(df_room_type_boston.columns):
    col = c.replace(' ','_')
    # append 'property_type' to each column name
    col = 'property_type_'+col
    # add to new columns list
    new_columns.append(col)

# set new column names
df_room_type_boston.columns = new_columns

# consolidate and create single df with variables of interest and price
df_boston = pd.concat([df_boston_listings[['price','bedrooms','bathrooms','comment_polarity']],df_neighbourhood_boston,df_room_type_boston], axis = 1)
# remove nas
df_boston = df_boston.dropna()


# create X and Y Dfs
X = df_boston
X = X.drop(['price'],axis=1)
Y = df_boston['price']


# split training and testing data
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.25, random_state=42)

# inti model
lm = LinearRegression(normalize=True)
# fit model
lm.fit(x_train,y_train)

# predict testing values
y_predict = lm.predict(x_test)
# get r squared
test_score = r2_score(y_test,y_predict)

# print model score on test data
#print(lm.score(x_test,y_test))
print("R2 Score (test): ", test_score)
print("Intercept: ", lm.intercept_)
```

<strong>Model Review</strong>
<br>
<ul>
    <li>R2 Score: Our model has an R2 value of .657, which means we can explain roughly 66% of the variation in listing prices within Boston. The current explanatory power of the model leaves much to be desired, but it provides a good frame of reference to analyze any new listings and determine if it presents a deal or not.</li>
    <li>Intercept: The intercept of 30.5 suggest that the baseline rate for a listing in Botson starts around $31, then prices move up or down according to the location, room type, reveiw sentiment, etc.</li>
</ul>

Below we dig into the individual model coefficients.

```python
# The following code is copied from an incalss Udacity example. It provides both the coefficient value and its aboslute value
def coef_weights(coefficients,x_train):
    '''
    INPUT: 
    Coefficients - Coefficients created by the linear model
    x_col - Columns from the x_df training set (ie the explanatory variables)
    
    OUTPUT:
    coefs_df - Returns a DF with the coefficient name, estimate, and absolute vlaue
    '''
    
    coefs_df = pd.DataFrame()
    coefs_df['est_int'] = x_train.columns
    coefs_df['coefs'] = coefficients
    coefs_df['abs_coefs'] = np.abs(coefficients)
    
    # sort in descending order
    coefs_df = coefs_df.sort_values('abs_coefs', ascending=False)
    return coefs_df

# Call newly created function
coefs_df = coef_weights(lm.coef_,x_train)
coefs_df
```

<strong>Model Coefficient Assessment</strong>
<br>
As a reminder, we undertook the price modeling by encoding different variables of interest (such as room type) and converted categorical variable of K options, to a binary matrix of K-1 features. With that in mind, the overall coefficient values make sense, however,  without any familiarty with the city of Boston it's a bit difficult to 'gut' check the model. That said, factors such as room type align with intuition as highlighted by -82.265675 value for 'shared room' and -66.077457 for 'private room'. That means, compared to 'Entire Home/Apt', assuming all other variables equal, a shared room in the same location and property type would go for 82 dollars less, while a private room, would go for 66 dollars less.  Thinking about this as a consumer, this make sense as generally speaking people are willing to pay more for an entire house/apartment, and pay less for simply a room in a house/apartment, and even less for shared room.

<br>

For variables such as 'bathrooms' and 'bedrooms', the coefficients mean that the price increases roughly 31 dollars  and 58 dollars all things equal, for each additional bathroom and bedroom, respectively. 

<strong>Average Listing Price by Room Type by Month</strong>
> Reviewing the previous charts, its clear that listing price is both seasonal and impacted by the type of the unit avaliable. To better understand the behavior of average listing price, the below chart segments by room type and then views listing price across time.

```python
# merge listing and calendar dataframes on listing id
df_listing_price_type_eom = pd.merge(df_boston_calendar[['listing_id','available','price','eomonth']],df_boston_listings[['id','room_type']],how='inner',left_on = 'listing_id', right_on = 'id' )
# filter to only show items marked as availiable == T
df_listing_price_type_eom = df_listing_price_type_eom[df_listing_price_type_eom.available == 't']
# create facet grid for each room type
g = sns.FacetGrid(data = df_listing_price_type_eom, col='room_type',size=6,sharey = False)
# plot bar chart by each month
g.map(sns.barplot,'eomonth','price');
# init labels list
labels = []
# loop through each created axes, rotates the x ticks 90 degress
for i, ax in enumerate(g.fig.axes):
    labels = [i.get_text() for i in ax.get_xticklabels()]
    for i in range(len(labels)):
        # split out UTC time data
        strDate = labels[i].split('T')[0]
        labels[i] = strDate
    ax.set_xticklabels(labels, rotation=90)
```
